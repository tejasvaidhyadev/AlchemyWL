```python
def get_ex(path, epoch):
    with open(path, 'r') as f:
        lines = f.readlines()
    ex_ls_incor = []
    ex_ls_cor = []
    fail_dict = {}
    succ_dict = {}
    ep_flag = 0
    for i in range(len(lines)):
        if lines[i] == "Epoch: " + str(epoch) + "\n":
            ep_flag = 1
#             print(lines[i])
        elif lines[i][:7] == "Epoch: ":
            ep_flag = 0
#             print(lines[i])
        if ep_flag == 1 and lines[i][:9] == "Example: ":
            ex_num = int(lines[i][9:-1])
            ex_ip = lines[i+1][8:-1]
            ex_op = lines[i+2][8:-1]
            ex_pred = lines[i+3][11:-1]
            ex_res = int(lines[i+4][8:-1])
            func = ex_ip.split("SEPARATE")[1].strip().split()[0]
#             print(ex_num, ex_ip, ex_op, ex_pred, ex_res)
            if ex_res == 0:
                ex_ls_incor.append([ex_ip, ex_op, ex_pred, ex_res])
                if func in fail_dict:
                    fail_dict[func] += 1
                else:
                    fail_dict[func] = 1
            else:
                ex_ls_cor.append([ex_ip, ex_op, ex_pred, ex_res])
                if func in succ_dict:
                    succ_dict[func] += 1
                else:
                    succ_dict[func] = 1
                
    print(len(ex_ls_incor))
    print(len(ex_ls_cor))
    print("Failures: ")
    print(fail_dict)
    print("Successes: ")
    print(succ_dict)
    return ex_ls_incor, ex_ls_cor
```
